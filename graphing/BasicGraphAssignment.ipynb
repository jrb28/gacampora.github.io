# Opiods VA

*Do opioid overdoes tend to be associated with less affluent areasâ€”that is, areas where families have lower incomes?*


```python
opiodsVA=pd.read_csv('OpidsVA.csv') #read in the data file

""" I will use a scatter plot to answer this question because it will show how overdoses compare to te median
household income, and you can detect any trends based on where the data points lie."""

fig, ax = plt.subplots() #create the figure
ax.scatter(opiodsVA['MedianHouseholdIncome'], opiodsVA['FPOO-Rate'], alpha=0.5, c = 'b') #scatter plot the data
ax.set_title('Opiod Overdoses vs. Median Household Income Levels') #Title the graph
ax.xaxis.set_label_text('Median Household Income') #Title the axis
ax.yaxis.set_label_text('Opiod Overdoses per 100,000 County Residents') #Title the axis
```

*Some people who start with opioid addictions are reported to transition to heroin use. What is the relationship in Virginia counties between opioid overdoses and heroin overdoses?*

```python
#using the same opiodsVA data

"""Similarly to the graph above, I choose to use a scatter plot because it shows how opiod overdoses compare to
heroine overdoses, and you can see the trends based on where the data lies."""

fig, ax = plt.subplots() #create the figure
ax.scatter(opiodsVA['FFHO-Rate'], opiodsVA['FPOO-Rate'], alpha=0.5, c = 'r') #create a scatter plot
ax.set_title('Opiod vs. Heroin Overdoses') #Title the graph
ax.xaxis.set_label_text('Heroine Overdoses per 100,000 County Residents') #Title the axis
ax.yaxis.set_label_text('Opiod Overdoses per 100,000 County Residents') #Title the axis
```

# Presidents

*Which states are associated with the greatest number of United States presidents in terms of the presidentsâ€™ 
birthplaces?*

```python
presidents=pd.read_csv('presidents.csv') #read in the data

"""For this question, I chose to use a histogram because it shows us the frequency of each state having a president,
so you can look at the highest bars to see which states had a president the most frequently."""

states=presidents['State'] #create a list of the different states
states=dict(states.value_counts()) #create a dictionary with the keys=state names and values=frequencies


"""We are asked for the states associated with the greatest number of US presidents, so I exclude
the states that have a frequency of one to make the graph simpler. If we want to look at all of the states,
we can exclude [:7] from the x and freq commands"""
x = list(states.keys())[:7] #X-axis data: the top 7 state names
freq=list(states.values())[:7] #Y-axis data: the top 7 state frequencies

fig, ax = plt.subplots() #create the figure
fig.set_size_inches(10,7) #make a bigger chart for clarity
ax.bar(x,freq, color='c', alpha=0.75) #create a bar graph using frequencies (histogram)

fig.suptitle('Presidents from Each State Frequency Histogram') #Title the graph
ax.xaxis.set_label_text('State') #Title the axis
ax.yaxis.set_label_text('Frequency of Presidencies') #Title the axis
```
